```python
import requests

url = "https://ecbc.in/datafile/Location_CZ_Latitude.json"
output_file = "Location_CZ_Latitude.json"

response = requests.get(url)

# Check if request was successful
response.raise_for_status()

# Save JSON file
with open(output_file, "wb") as f:
    f.write(response.content)

print("JSON file downloaded successfully.")
```

```python
import json

input_file = "Location_CZ_Latitude.json"
output_file = "Location_CZ_Latitude_cleaned.json"

with open(input_file, "r", encoding="utf-8") as f:
    data = json.load(f)

cleaned_data = []

for entry in data:
    climate_zone = entry.get("Climate_Zone", "")

    # âœ… REMOVE ONLY if comma is present
    if "," in climate_zone:
        continue

    cleaned_data.append(entry)

with open(output_file, "w", encoding="utf-8") as f:
    json.dump(cleaned_data, f, indent=2, ensure_ascii=False)

print(f"Cleaned file saved as: {output_file}")
```
